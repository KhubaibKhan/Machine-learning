# Logistic Regression
        Logistic Regression is used to predict descrete value output, like if you want to predict the digit from an image.
        Logistic Regression is an example of Supervised learning because we have give it labeled data, then ask it to label the unlabled data.
# Hypothesis: 
    We can use the hypothesis of linear regression for logistic which will fit a straight line that will classify the data but it will only work for simple data. Using the hypothesis of linear regression for Logistic regression is not a good idea.
## Logistic Regression Model:
    In logistic regression we want the hypothesis to output a value between o  and 1.
$$ 0 <= h_Q(x) <= 1 $$
    For linear Regression we had.
$$ h_Q(x) = Q^Tx $$
    For Logistic Regression we will use a slightly modified version.
$$ h_Q(x) = g(Q^Tx) $$
where $$ g(z) = \frac{1}{1 + e^{-Q^Tx}} $$
This function is known as Sigmoid or Logistic function. This is where Logistic Regression got its name from.

```python
import numpy as np
x = np.linspace(-10, 10, 100)
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
y = sigmoid(x)
```

    The sigmoid function will always gives value between 0 and 1.

## Logistic Regression Cost Function:
     The cost function is difference between the original value and the predicted value. In linear regression we had used RMSE cost function. In logistic regression we will chose another cost function.
     In losgistic regression we use different cost function because for logistic regression the linear regression cost is not convex which is very hard to minimize.
     
$$ Cost(h_Q(x), y) = { {-\log(h_Q(x)) if y=0}......\&....... 
{-\log(1 - h_Q(x) if y=1}}$$
* $Cost = 0$ if $y=1$ and $ h_Q(x)=1 $ 
* $Cost = 0$ if $y=0$ and $ h_Q(x)=0 $
* $Cost = \infty$ if $y=0$ and $ h_Q(x)=1 $ or if $y=1$ and $ h_Q(x)=0 $
        The below graph prove the above points.

```python
def cost(y_pred, y_true):
    cost = np.mean((y_true*np.log(y_pred) + (1 - y_true)*np.log(1 - y_pred)))*-1
    print(cost)
    return cost
```

 ### Hypothesis:

\begin{align}
\dot h_x & = g(z).....z = \theta_0 + \theta_1X
\end{align}
### Parameter:

$$ \theta_0 and \theta_1 $$
### Cost Function:

$$ J(\theta) = -\frac{1}{m}[ \sum_{i=1}^{m}{y^{(i)}\log(h_Q(x^{(i)}))}
{+(1 - y^{(i)})\log(1 - h_Q(x^{(i)})}]$$

### Goal: 

$$ Minimize  \ J(\theta_0, \theta_1)  using  \theta_0 and \theta_1 $$

## Gradient Descent:
Gradient descent is just an algorithm that minimize some arbitrary function $ \ J(\theta_0, \theta_1) $. There could be many $\theta $ not just two like $ J(\theta_0, \theta_1, \theta_2, \theta3, ....) $
### Outline:
* Start with some $\theta_0 and \theta_1 $.
* keep changing $\theta_0 and \theta_1 $ to reduce $ J(\theta_0, \theta_1)$ untill hopefully we endup at a minimum.
### Algorithm:
    The algrithm of Gradient Desent.
Repeat until convergence {

$$ \theta_j := \theta_j - \alpha \frac{\mathrm{d}}{\mathrm{d}\theta_j}J(\theta_0, \theta_j) $$
}


$$ \theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta (x^{(i)}) - y^{(i)}) x^{i}_j $$

    Now that we have the Hypothesis, cost function and gradient descent algorithm lets implement it using python from scratch.

```python
class MyLogisticRegression():
    def __init__(self, no_of_attrs, alpha = 0.2, iteration=1000):
        self.alpha = alpha
        self.theta = np.random.rand(no_of_attrs)
        self.t_cost = []
        self.iter = iteration
        self.b = 0
    def fit(self, x_train, y_train):
        if len(self.theta) == 1:
            self.x = x_train.reshape(len(x_train), 1)
        else:
            self.x =  x_train
        for i in range(self.iter):
            Q_cost = self.cost(y_train)
            self.t_cost.append(Q_cost)
            self.GradientDescent(y_train)
            if i%1000 == 0:
                print(i, "Epoch completed with cost:", Q_cost, "theta", self.theta)
        print(10000, "Epoch completed with cost:", Q_cost, "theta", self.theta)
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def hypothesis(self, x_value):
        return self.sigmoid(np.dot(x_value, self.theta) + self.b)
    def cost(self, y_train):
        cost = np.mean((y_train*np.log(self.hypothesis(self.x)) + (1 - y_train)*np.log(1 - self.hypothesis(self.x))))*-1
        return cost
    def cost_der(self, y_given):
        cost = (self.hypothesis(self.x) - y_given)
        cost0 = cost
        cost = cost*self.x.T
        cost = np.mean(cost)
        cost0 = np.mean(cost0)
        return (cost0, cost)
    def GradientDescent(self, y_given):
        (gr0, gr1) = self.cost_der(y_given)
        self.theta = self.theta - self.alpha*gr1
        self.b = self.b - gr0
        
    def predict(self, x_test):
        return np.round(self.sigmoid(np.dot(x_test.reshape(len(x_test), len(self.theta)), self.theta) + b))
    def weights(self):
        return (self.theta, self.b)
```

```python
x_train = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [6, 7], [7, 8]])
y_train = np.array([1, 1, 1, 0, 0, 0])
log_reg = MyLogisticRegression(no_of_attrs = 2, iteration=10000)
log_reg.fit(x_train, y_train)
```

### Making sure the model is working fine:
* If $\ J(\theta)$ is not converging it means gradient descent is not working. Try changing the $\alpha$.
* If $\alpha$ is too small, gradient descent can be slow to converge.
* I $\alpha$ is too large $\ J(\theta)$ may not decrease and the model will not converge.
* Chose alpha between $ 0.001 and 0.3$. It depends on the problem mostly.

## Logistic Regression using scikit-learn:
    We can implement the logistic regression using a few lines of code.
    We will classify the flower data which is already present in sklearn.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x, y = load_iris(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
log_reg = LogisticRegression(random_state=0,solver='lbfgs', multi_class='multinomial', max_iter=100).fit(x_train, y_train)
```

```python
log_reg.score(x_test, y_test)
```

    The model gives about 98% accuracy on the test set so it is working fine.

## Logistic Regression using Tensorflow:

```python
import tensorflow as tf
```

```python
NUM_STEPS = 1000

x_data = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [6, 7], [7, 8]])
y_data = np.array([1, 1, 1, 0, 0, 0])
#y_data = y_data_pre_noise + y_data
wb_ = []
loss_ = []

print(np.shape(x_data))

g = tf.Graph()
with g.as_default():
    x = tf.placeholder(tf.float32)
    y_true = tf.placeholder(tf.float32)
    with tf.name_scope('inference') as scope:
        w = tf.Variable([[0, 0]], dtype=tf.float32, name='weights')
        b = tf.Variable(0, dtype=tf.float32, name='bias')
        y_pred = tf.matmul(w, tf.transpose(x)) + b
    with tf.name_scope('loss') as scope:
        loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=y_true, logits=y_pred)
        loss = tf.reduce_mean(loss)
    with tf.name_scope('train') as scope:
        learning_rate = 0.1
        optimizer = tf.train.GradientDescentOptimizer(learning_rate)
        train = optimizer.minimize(loss)
    init = tf.global_variables_initializer()
    with tf.Session() as sess:
        sess.run(init)
        for step in range(NUM_STEPS):
            sess.run(train, {x:x_data, y_true:y_data})
            if step%100 == 0:
                print(step, sess.run([w, b]), sess.run(loss, {x:x_data, y_true:y_data}))
                wb_.append(sess.run([w, b]))
            loss_.append(sess.run(loss, {x:x_data, y_true:y_data}))
        print(1000, sess.run([w, b]), sess.run(loss, {x:x_data, y_true:y_data}))
```

```python
trace = go.Scatter(
    x = np.arange(1000),
    y = loss_,
    mode = 'lines',
    marker = dict(
        color = 'red'
    )
)
layout = go.Layout(
    title = "Cost per iteration",
    titlefont = dict(
        color = "rgba(200, 200, 200, 200)"
    ),
    xaxis = dict(
        title = "No of iteration",
        color = "rgba(200, 200, 200, 200)"
    ),
    yaxis = dict(
        title = "Loss",
        color = "rgba(200, 200, 200, 200)"
    ),
    paper_bgcolor='rgba(1,1,1,0.5)',
    plot_bgcolor='rgba(0,0,0,0)'
)

fig = go.Figure(data = trace, layout = layout)
fig.show()
```
